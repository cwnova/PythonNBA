Chapter 3: Object Oriented Programming

````python
class Player:

    def __init__(self, name, pos, fgm, fga, threes_m, threes_a):
        self.name = name
        self.pos = pos
        self.fgm = fgm
        self.fga = fga
        self.threes_m = threes_m
        self.threes_a = threes_a

    def fg_percent(self):
        return self.fgm / self.fga
    
    def threes_percent(self):
        return self.threes_m / self.threes_a
   
    def twos_percent(self):
        return (self.fgm - self.threes_m) / (self.fga - self.threes_a)
    
    def shooting_percent(self):
        return {
          'overall': self.fg_percent(),
          '3P%': self.threes_percent(),
          '2P%': self.twos_percent()
        }

curry = Player('Steph Curry', 'PG', 535,1224, 285, 750)

print(curry.pos)

print(round(curry.fg_percent(), 3))

print(curry.shooting_percent())
````

To tell the truth, anonymous functions are a bit more higher-level of a topic but we're going to need it when we move on to using pandas.

We went over user-defined functions using the keyword def. Defining functions that way is great because we write our own function and then we can use it over and over again throughout the code.

Sometimes we need to write a function and use it just once. This usecase typically occurs when we need to pass our function as an argument to another function. This can be mind warping for beginners, but remember functions are objects just like anything else. When we start to use pandas, we will be using a pandas method called apply that will take a function as the argument, and we will define the function in one line right in the parentheses, using something called an anonymous or lambda function.

We can't go over apply just yet as we haven't gotten to using pandas, so I want to introduce you to map , a built in function that allows us to "map" a function over an iterable. It won't be useful to you forever, as list comprehensions essentially fulfill the same job in a much more legible manner, but it's useful to know in case you need it.


````python
baskets_made = [3, 11, 7, 8]

def mul_by_two(x):
    return x*2

    #map allows us to 'map' a function over an iterable

#What map does is it "maps" over a list and runs a function across that list. 
#It returns something called a map object which is iterable but not able to be read directly. 
#We'd need to covert it to a list again to be able to read the items within it as a list.

points_scored = map(mul_by_two, baskets_made)   ##tells to use mul_by_two function with baskets made dictionary
points_scored = list(points_scored)   #assigns point_scored the list of points_scored assigned above

print(points_scored)
````

````python
#same as above but using lambda
#The syntax to write a lambda function is simply lambda argument: return_value_modified . 
#First we specify our argument, then we specify how we want to manipulate that argument.

baskets_made = [3, 11, 7, 8]

points_scored = map(lambda x: x*2, baskets_made)
points_scored = list(points_scored)

print(points_scored)
````

Some people, usually non-experienced Python programmers, will do something like this:



```
# from pandas import *
```


What this does is import all Python objects from the pandas module and insert them into the current modules namespace. This is known, colloquially, as polluting the namespace. The namespace is the objects that are able to be called within the environment you're working from.

To pollute the namespace like this is bad, bad code. It won't always get you an error, but it's bad style. I'm willing to break convention every once in a while, but this is just bad. When writing Python code, you want to be as explicit as possible where an object came from. It's one of the guiding principles of Python code, actually; In Python, explicit is better than implicit.

Explicitly makes code much more legible when you can tell where specific objects came from. For example, if I were to say import pandas as pd, whenever I wanted to use an object from the pandas library I'd have to use pd.DataFrame() , for example. The pd in front of DataFrame here tells myself and who ever is reading my code that this DataFrame class came from the pandas module imported earlier. If I were to do from pandas import * , then I could just reference DataFrame anywhere in my code, without any reference to pd or pandas. This makes code really hard to follow and can lead to some unusual errors (for example, let's say there's a function called pivot_table within the pandas module, what would happen if I then wrote my own function called pivot_table in my own module?)

Similarly, import numpy as np is better style than from numpy import linspace . Why? Because I could easily reference the function from numpy's module linspace by using np.linspace. The latter option is still acceptable, certainly more acceptable than doing import * , but less explicit. If you can be more explicit about where you're accessing objects from external modules, do it.

**Problem Solving using Programming**

Now that we have EFF by team we want to gauge if the teams with a higher EFF performed better. One way to do this is look at the number of wins a team got. This is a pretty basic but often effective way to gauge a team's performance. So the first goal is to create a DataFrame that has wins for each team over the 2021-22 season. We do this by first pulling the data for TeamGameLogs from the nba_api .

Interacting with the NBA API can be tricky at times, and we'll need to pull in yet another library (but one that exists within Python, thus no installation is necessary) called json . You can view the documentation for the json module here. This module helps us work with data in JSON format, which is essentially a data format that accomplishes two things: (A), it's easy for humans to read, write and decipher, and it's (B) easy for computers to work with. You can read more about JSON here, but I encourage you to do some independent research on JSON as well, as this won't be the first time you interact with JSON in your programming career.

Here is a more detailed explanation of each line:

***import json*** imports the json module, which provides functionality for working with JSON data.

***response*** = endpoints.teamgamelogs.TeamGameLogs(season_nullable = '2021-22') invokes an API endpoint called TeamGameLogs from the endpoints.teamgamelogs module, and specifies the season_nullable parameter as '2021-22' . This sends a request to the API for data from the 2021-22 season.

***content*** = json.loads(response.get_json()) parses the JSON response from the API into a Python dictionary using the json.loads() function. You can read more about this function here. The response.get_json() method is used to retrieve the JSON response as a string.

***results*** = content['resultSets'][0] accesses the first element of the resultSets key in the content dictionary. This key typically contains a list of dictionaries with data from the API.

***headers*** = results['headers'] retrieves the list of column names from the headers key in the results dictionary.

***rows*** = results['rowSet'] retrieves the data rows from the rowSet key in the results dictionary.

***record*** = pd.DataFrame(rows, columns=headers) creates a Pandas DataFrame from the rows data and assigns the column names from the headers list.

***record.head()*** displays the first few rows of the DataFrame using the head method.


````python
import json

response = endpoints.teamgamelogs.TeamGameLogs(season_nullable = '2021-22')

content = json.loads(response.get_json())
results = content['resultSets'][0]
headers = results['headers']
rows = results['rowSet']

record = pd.DataFrame(rows, columns=headers)

record.head()
````

**Value Counts**

It looks like our DataFrame includes every game from the 2021-22 season and in the WL column it tells us the outcome of the game. This column is key in calculating the number of wins for each team! 

I recommend trying to figure out how to transform this DataFrame into a DataFrame that contains the following: team name, number of wins over the course of the season. 

To do this we will use a groupby like we did earlier since we want to group by each team and add up the number of wins they got. 

Since the WL column isn't numeric, we cant actually add up the number of wins, instead we can use python to count the instances of W and L in the column. We do this with a value_counts method. It counts the number of times each value shows up in a given column.

````python
record = record.groupby(['TEAM_ABBREVIATION'])['WL'].value_counts()
record = pd.DataFrame(record).rename(columns = {'WL':'W'})  ##count the number of Ws in WL column
record.head()
````

**Pivot Tables**

A pivot table allows for reformatting a DataFrame's columns and indexes. In our case, we want wins and losses to be two separate columns instead of being in one column.

````python
record = pd.pivot_table(record,values = 'W', index = 'TEAM_ABBREVIATION', columns = 'WL')
record.head()
````

 A merge is done to join two separate DataFrames into one DataFrame. I recommend reading a bit further on merges since they take a little bit of work to understand and are a very important concept in working with data.

**Joins and Merges in DataFrames**

You might come across pandas joins as well and I'll explain the difference now. Joins and Merges accomplish the same thing, combining two DataFrames into one DataFrame, but have subtle differences. The biggest is that a join combines the DataFrames on their index and with a merge we have to be explicit and tell pandas what we want the DataFrames to merge on. 

I almost exclusively use merges since it is better to be explicit and tell Python how you want something done. Another benefit of using marges is you don't have to deal with indexes, which adds another layer of complexity.

If you think about two DataFrames being combined, you have to consider how they are being combined. For example, we want to combine our EFF DataFrame and our wins DataFrame. What is common between the two DataFrames? The team name. So we want to join on the team name, this means each row will match up team, and include the rest of the information from both DataFrames.

When merging DataFrames it's important to think about it visually, one DataFrame on the left, and one DataFrame on the right joining up row by row. Often times the left and right DataFrame do not have exactly the same indexes (they do in our case so we don't need to worry about this) so you have to specify the type of merge. The main types of merges include: inner, outer, and left merge. I recommend reading further on this as I am not going to get into how these are different, but it is important to understand how your DataFrames are being glued together. Here is a Stack Overflow thread that was meant as an SQL question but applies in our case here as well. https://stackoverflow.com/questions/38549/what-is-the-difference-between-inner-join-and-outer-join

When we actually run the merge line, like I mentioned before, we have to be explicit about the columns we are merging "on" - the columns we are telling Pandas to match up for us. In our case we want to merge on team which means our left DataFrame (the EFF DataFrame) will merge on TEAM and the right DataFrame (the wins DataFrame) will merge on TEAM_ABBREVIATION . We also could have changed the column names so that they match and simply used a singular on clause joining on the matching name, but in the early going it is crucial to understand what is meant by the left and right DataFrame when it comes to merges.

````python
team_eff = team_eff.merge(record, left_on = 'TEAM', right_on = 'TEAM_ABBREVIATION')

team_eff.head()
````

**Residual Plots**

Going back to the regression plot, seaborn also has capabilities to plot residual plots, which essentially show us the difference between our regression model (the line of best fit), and the points on the x/y plane. This is especially important for data science and what we'll be doing in the machine learning sections.

In the ML sections, we'll be fitting a line to some data. The more "bunched" together the data we're fitting the line to is around the line of best fit, the better our line of best fit is actually fit.

We are then going to take this line of best fit and use it to predict new values. These new values then get compared to real-world values and the difference between the predicted value and then real-world value is called the residual.

Analyzing the residuals of your model is an extremely important facet of modeling, and seaborn provides a quick helper function to plot the residuals for a line of best fit between two variables.

````python
sns.set_style('dark') # changing up the style

##sns.residplot(season['FGA'],season['PTS'])
sns.residplot(data=season, x="FGA", y="PTS")

plt.title('Residual plot')
plt.xlabel('FGA') # this is how we set an x label in matplotlib
plt.ylabel('PTS') # this is how we set a y label in matplotlib
plt.show() # show our plot
````

Points bunched around the 0 part of our line indicate that our line of best fit perfectly matched up with the an X, Y point on the plane.

The x-axis represents our independent variable (or our input into the model), while the y-axis represents our residual. To reiterate, residual is how wrong our line of best fit was. In short, it's our error. The farther away a point is from the y = 0 line, the bigger our error.

You can see as we got further and further away from 0, the distribution of our residual around the y axis increased. This means that it got harder and harder for our line of best fit to predict values as field goals attempted. When looking at residual plots, you want to ensure that the distribution of your residuals are relatively symmetrical along both axis's. If they aren't, then your model has room for improvement.

Points above the y-axis represent those points that our line of best fit overestimated a player's points scored. Points below represent those points that we underestimated. As you can see, the distribution along the y-axis is relatively equally distributed, but as we go along the x-axis, it does widen. This indicates that further out in the player's with high number field goals attempted, there is something else that our model is missing that can help predict points scored (how efficient their shooting is). For those players with a low amount of field goals, our model can accurately forecast their performance.

**Drawing a Court**
Next step after pulling the shot chart data is to have a court to plot the data onto. We'll do this by hand just to show its possible, but don't worry too much if you don't follow every line below. Essentially we are physically drawing each part of the court, one line at a time, and stitching it all together into a diagram of a court.

````python
from matplotlib.patches import Circle, Rectangle, Arc

def draw_court(ax=None, color='black', lw=2, outer_lines=False):
    # If an axes object isn't provided to plot onto, just get current one
    if ax is None:
        ax = plt.gca()

    # Drawing the various parts of an NBA basketball court

    # Create the basketball hoop
    # Diameter of a hoop is 18" so it has a radius of 9", which is a value
    # 7.5 in our coordinate system
    hoop = Circle((0, 0), radius=7.5, linewidth=lw, color=color, fill=False)

    # Create backboard
    backboard = Rectangle((-30, -7.5), 60, -1, linewidth=lw, color=color)

    # The paint:
    # Create the outer box 0f the paint, width=16ft, height=19ft
    outer_box = Rectangle((-80, -47.5), 160, 190, linewidth=lw, color=color,
                          fill=False)
    # Create the inner box of the paint, widt=12ft, height=19ft
    inner_box = Rectangle((-60, -47.5), 120, 190, linewidth=lw, color=color,
                          fill=False)

    # Create free throw top arc
    top_free_throw = Arc((0, 142.5), 120, 120, theta1=0, theta2=180,
                         linewidth=lw, color=color, fill=False)
    # Create free throw bottom arc
    bottom_free_throw = Arc((0, 142.5), 120, 120, theta1=180, theta2=0,
                            linewidth=lw, color=color, linestyle='dashed')
    # Restricted Zone, it is an arc with 4ft radius from center of the hoop
    restricted = Arc((0, 0), 80, 80, theta1=0, theta2=180, linewidth=lw,
                     color=color)

    # Three point line
    # Create the side 3pt lines, they are 14ft long before they begin to arc
    corner_three_a = Rectangle((-220, -47.5), 0, 140, linewidth=lw,
                               color=color)
    corner_three_b = Rectangle((220, -47.5), 0, 140, linewidth=lw, color=color)
    # 3pt arc - center of arc will be the hoop, arc is 23'9" away from hoop
    # I just played around with the theta values until they lined up with the 
    # threes
    three_arc = Arc((0, 0), 475, 475, theta1=22, theta2=158, linewidth=lw,
                    color=color)

    # Center Court
    center_outer_arc = Arc((0, 422.5), 120, 120, theta1=180, theta2=0,
                           linewidth=lw, color=color)
    center_inner_arc = Arc((0, 422.5), 40, 40, theta1=180, theta2=0,
                           linewidth=lw, color=color)

    # List of the court elements to be plotted onto the axes
    court_elements = [hoop, backboard, outer_box, inner_box, top_free_throw,
                      bottom_free_throw, restricted, corner_three_a,
                      corner_three_b, three_arc, center_outer_arc,
                      center_inner_arc]

    if outer_lines:
        # Draw the half court line, baseline and side out bound lines
        outer_lines = Rectangle((-250, -47.5), 500, 470, linewidth=lw,
                                color=color, fill=False)
        court_elements.append(outer_lines)

    # Add the court elements onto the axes
    for element in court_elements:
        ax.add_patch(element)

    return ax
````

Now we'll begin to talk about what machine learning is and how we can use it to make predictions in basketball.

Machine learning algorithms are algorithms that build models on data based on sample data, or training data, and can then be used to predict or make decisions based off new data, or testing data.

**Machine Learning Algorithms**
Machine learning algorithms fall into two categories, either supervised or unsupervised. Supervised learning algorithms learn with labeled data, while unsupervised learners learn without reference to labels (more on these in Section 11!)

**Supervised Algorithms**
A supervised algorithm is trained on data that is labeled, and then used to predict labels for unlabeled data. In regression, "labels" is in reference to some continuous output.

Supervised algorithms themselves fall into two main categories: classification and regression.

Classification algorithms (K-Nearest Neighbor, K-Means Clustering, Support Vector Machines, Naive Bayes, Logistic Regression, Random Forests, Decision Tree) aim to find the class to which a set of inputs belongs to. For example, predicting the position an NBA player plays based off height, weight, rebounds and assists is a classification problem.

This section focuses on regression algorithms. As stated above, regression algorithms aim to predict a continuous output. A continuous output is an output that can span any interval on the number line. Most of the data we work with in basketball is continuous, such as: number of points, at rebounds, blocks, and shooting percentages.

**How are We Using Algorithms?**
In the case of this course module, we will be training our algorithm on data going back to 2000 for past player season statistics to predict the number of points a player will score in the next season. Note that since this is an autoregressive model (a model that uses past behavior to explain future behavior), rookies and players with limited playing time in the prior season will not be included in the mix.

I am going to go a bit into the math of linear regression. This isn't crucial to be able to do linear regression, but it is nice to understand the basics.

**Linear Regression Math**

Linear regression models "learn" by estimating coefficients and an intercept for a line of best fit for one feature (for 2 features, the regression model finds the plane that best fits the training data, for 3 features or more, it fits a hyper-plane). These learned coefficients and intercept are known as model parameters, and they are learned through the training process.

This learning process of "finding the best fit" is achieved by minimizing the sum of the squared residuals between the training data and the line of best fit.

As you can see here, the model adjusts a line of best fit, until the summed squared distance of the red lines (the residuals; a residual is the vertical distance between a training point and the line) is minimized. You can think of the algorithm as tweaking the line of best fit by adjusting the intercept and the slope (hello algebra) many times and keeping track of the summed squared of the residuals each time, until it has found a minimum value. In this image here, b0 (intercept) and b1 (the slope) are the model parameters that are learned through the training process. This process of tweaking the line (our model) can be expanded to more than one feature as well. The key point here is that model is estimating the coefficients and intercepts that minimizes the overall sum of the squared residuals.

Also, residuals are squared so as to keep all values positive. The absolute value of the residuals could also work, but mathematically, squared values are easy to work with.

Model hyperparameters, which we won't cover in this section, are not learned through the training process, but are tuned through another process called cross validation.

````python
import pandas as pd
import numpy as np
from sklearn import linear_model
import requests
from nba_api.stats import endpoints
from matplotlib import pyplot as plt

pd.set_option('display.max_columns', 100)
````

To load in our data I am using the **leagueleaders** endpoint from our nba_api. The way this endpoint works is first we specify which season of data we would like to pull. Because we want multiple seasons of data we have to pull each season individually and then put them together into a singular DataFrame. The way you combine DataFrames like this is with the **append** method. This essentially takes one DataFrame and tacks on its rows to the bottom of your other DataFrame.

Putting all of this together we are going to make a **for** loop that takes in a season of data and then appends it to the previous season, and continue this loop until we have 20 seasons worth of information. The tricky part is the syntax of the endpoint requires us to write the year in a format like **YYYY-YY **since an NBA season is across two calendar years. We could write out each season individually but let's see if we can utilize python to do this for us, instead.

````python
'''
Organizing the Data
I am going to make two lists, the first list will represent the first part of the year format, and the second list will represent the second part of the year list.
'''

seasons = np.arange(2000,2021)
seasons2 = np.arange(0,22)

print(seasons2)
````

````python
'''
Now we just need to put these two lists together. We can do this using the format method which allows us to insert values into a string. 
So we'll loop through the list, take a value from the first and a value from the second, put it together, and pull data from that season.
'''

data = endpoints.leagueleaders.LeagueLeaders(season='1999-00').league_leaders.get_data_frame()
data['season_start'] = 1999

for i, v in enumerate(seasons):
    temp = endpoints.leagueleaders.LeagueLeaders(season='{0}-{1}'.format(v,str(seasons2[i]+1).zfill(2))).league_leaders.get_data_frame()
    temp['season_start'] = v
    data = data.append(temp)

df = data # this is just making a copy of our DataFrame by renaming it so we dont have to redownload all the data later
````

**Learning about Sklearn**
Finally, we bring in Python's de-facto machine learning library, **sklearn** . You can visit the documentation for scikit learn here.

Built on an extremely streamlined and clean API known as the estimator API, Scikit learn has a large host of machine learning algorithms, both supervised and unsupervised. It also comes with a host of helper objects for evaluating and tuning our models.

The estimator API is streamlined in the fact that every single supervised machine learning algorithm is implemented in about the same fashion.

**Split Data and Create Model**

Each algorithm is defined as a class. In the case of this section, we are going to be using the **LinearRegression** model, that comes from the **sklearn.linear_model** module.

That model is then instantiated with model hyperparameters. In this case, we have no model hyperparameters we set here, and so our model is instantiated with no arguments.

Then, you pass in **X** training data and **y** training data where **X** has the shape (n_samples, n_rows), and y is an array with the shape (n_samples, 1) to the **fit** method. The **fit** method is what trains the algorithm.


You can see above that we use a utility function to split our data into training and testing sets. This is vitally important. You do not want to test data on the same data that you trained it with; this is known as data-leakage, and can prevent you from accurately assessing the performance of your model.

Instead, you want to partition part of your **X** and part of your **y** into testing and training sets, and then train the model with **X_train** and the y_train data, use the x_test data to predict new values with the trained algorithm, and use y_test to evaluate your model score.

For our case, we use the train_test_split function to split our data. This function gives us back a 4-element tuple, which we can unpack by setting the variables **X_train** , **X_test** , **y_train** , **y_test** . We set the argument test_size=0.2 to train our algorithm with 80% of the data and test it with the remaining 20%. This isn't the only way to split up our data. Using K-Fold cross validation, which is out of the scope of this course, may be useful to take a look at for you in other cases.

Anyway, after training your algorithm using the fit method, you can predict new values using the predict method.

After finding a y_pred you can evaluate your model score by comparing it to y_test .

Here, we are using a utility function from sklearn.metrics to calculate the average absolute value of the distance between real-world values, and predicted values. This distance is our error, and so mean_absolute_error is the average absolute value of our model's error.

**Putting it in Practice**

Let's grab our data and split it up like mentioned above. It's really important you understand each piece of X_train, X_test, y_train, y_test.

First we take our original data and split it into train data (top right) and test data (bottom right). Both of the datasets are the same with the only difference being that training datasets are much bigger to put as much data as we can towards the learning step. Each of the orange columns is a feature in our X data. Our y data will always have one output, which in our case is next year points scored.

So that's the framework, using multiple columns to predict a single output. We do this two times: once for our training data and then once for our test data.

````python
from sklearn.model_selection import train_test_split

X = df[['PTS', 'FGM', 'FGA','FTM','FTA','EFF','TOV','MIN']].values
y = df[['PTS_Next_Year']].values

print("Original Data Shape - X: {0}, Y: {1}".format(X.shape,y.shape))

# split up our data into 20% testing, 80% training
# train_test_split documentation
# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)

print("Train Data Shape - X: {0}, Y: {1}".format(X_train.shape,y_train.shape))
print("Test Data Shape - X: {0}, Y: {1}".format(X_test.shape,y_test.shape))
````

Now we have our data formatted so that we have specific data to use for our training step and data set aside for the testing step.

**Make sure you don't touch the test data until you are done with your model!!** 
I want to reiterate this because this is the most common mistake in machine learning. 

The point of test data is that you want to pretend you don't know the outcome so that we can attempt to predict it and see how well we do. If you train on your test data, or even peek at your test data you are tampering with the training process and the results may be skewed based on the knowledge you get from it. It is common practice to set aside development data so that you can treat this as an extra set of test data that you can use to help develop your model down the line.

One reason why it is so important to make sure you don't touch the test data is in most instances data scientists have a limited amount of data and if the test data is wasted there won't be any left to test on. Lucky for us basketball has decades and decades of data so we can always grab more if you do make the mistake of training on your test data.

**Training and Testing Data**

Let's first initialize our linear regression model and train on our training data.



````python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

# sklearn.linear_model.LinearRegression documentation
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

lr = LinearRegression()

# train the algorithm
# the fit method documentation
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression.fit

lr.fit(X_train, y_train)
````

Now that our model did the fit step it is able to make predictions. The next step in the machine learning process is to check the errors on our training data. To do this we are going to predict on our X_train data and then compare our results to our Y_train data.

I know this may seem weird predicting on data we trained on, but this is just another step to allow you to go back and make modifications to your model before you test on your test data and are essentially done. The relationship between the error you get on your training data and the error you get on your test data is really important. When we look at the errors on our training data, the more complex we make our model (more features) and the lower our error will get. This is not necessarily good though because it means our model is learning our dataset too in-depth. This means it will not generalize to other data. This concept is called over fitting.

Imagine a model trained to write music that can only write the exact music it was trained on. This is an example of an over fitted model. It is completing what is asked of it, but too perfectly based on the information it was trained with.

The goal is to make our model complex enough where we can make good predictions but not too complex where it will not work on other datasets. By the nature of our process our test error will almost always be higher than our training error since we are learning our training data.

**Testing Errors**

As mentioned before, let's check the errors in our training data.

````python
"""
Predicted values based off testing data. We are going to compare these predicted values to 
real world values and try to quantify the difference between our model and reality
"""
# predict method documentation
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression.predict
y_pred = lr.predict(X_train)

# mean_absolute_error documentation
# https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_absolute_error.html
print('Mean number of Points:', y_train[:,0].mean())
print('Mean absolute error:', mean_absolute_error(y_pred, y_train))
````

Above we predicted on **X_train** using the predict method and checked the error between our predicted and the actual values which are y_train here. Training data is always matched with training data.

Here is a good point to stop and think about your model. Did it perform well? Do you want to try out different features to see if you can improve performance? Any tinkering you'd like to do needs to be done now because the next step is to perform on test data. And like I've mentioned many times, once you test on the test data your process is basically over.

**MVP Mini Example**

Let's start with a simple example for predicting the league MVP. A major indicator of this is the number of points scored by a player during the season. If we created a model to predict MVP solely based off points scored (its better to use more features but this would be harder to visualize) it would look something like this:


(Feel free to dig deep into the math of a sigmoid function, but it is not required to get an idea of how logistic regression works. It is important to take from the plot that we have our sigmoid function in a pale blue, and our individual players points scored in red.)

````python
import math 
from matplotlib import pyplot as plt
import numpy as np

def sigmoid_line(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp((-1*item))))
    return a
x = np.arange(-10., 10., .2)
sig = sigmoid_line(x)

points_scored = [1000,1400,1500,1800,2000]

plt.plot(50*x+1500,sig)
plt.plot([1000,1200,1500,2000],[0,0.005,0.5,1],'ro')
plt.xlabel('Points Scored in 2022')
plt.ylabel('MVP Probability')
plt.show()
````

**Categorizing our Results**  

For example, if a player scored 2000 points (the top right red dot) our sigmoid would return 1. Or vice versa, if a player scores 1000 points the sigmoid returns 0. It gets interesting when we look at the point in the middle, a player that scored 1500, since our sigmoid predicts 0.5. So up until now we just have the probabilities the function returned, but it is our job to convert these to have actual meaning. In most cases a logistic regression model will assume that above 0.5 returns the "yes" case (wins MVP) and below 0.5 returns the "no" case (does not win MVP). This threshold is actually arbitrary and can be moved around depending on the use case. We could also look directly at the probabilities the sigmoid function gives us, rather than converting it into a yes / no. Often you get more information out of looking at the probabilities.

**Normalization**

Data normalization is used in machine learning to make model training less sensitive to the scale of features. This allows our model to converge to better weights and, in turn, leads to a more accurate model. Normalization makes the features more consistent with each other, which allows the model to predict outputs more accurately.

Normalization is a rescaling of the data from the original range so that all values are within the new range of 0 and 1. This is done by dividing all values by the maximum value encountered or by subtracting the minimum value and dividing by the range between the maximum and minimum values.

Normalization Math
A value is normalized as follows:

- y = (x – min) / (max – min)

Where the minimum and maximum values pertain to the value x being normalized.

For example, for a dataset, we could guesstimate the min and max observable values as 30 and -10. We can then normalize any value, like 18.8, as follows:

- y = (x – min) / (max – min)

- y = (18.8 – (-10)) / (30 – (-10))

- y = 28.8 / 40

- y = 0.72

You can think of it as a rescaling of the data so that the model doesn't place more focus on larger numbers (ex. points scored) over smaller numbers (ex. blocks).

We apply this concept of normalization to our entire DataFrame on each season. **True and False Positives, True and False Negatives**

The report shows the main classification metrics precision, recall and f1-score on a per-class basis (we have two classes, MVP and no MVP). The metrics are calculated by using true and false positives, true and false negatives. Positive and negative in this case are generic names for the predicted classes. There are four ways to check if the predictions are right or wrong:

TN / True Negative: when a case was negative and predicted negative

TP / True Positive: when a case was positive and predicted positive

FN / False Negative: when a case was positive but predicted negative

FP / False Positive: when a case was negative but predicted positive

It is crucial to apply this knowledge on a case by case basis since these have a different level of importance based on the model application. For example, if we are getting a lot of false positives in our case, this means for some reason the model thinks that many players are capable of winning MVP, and this can't be true. A classic example used to get the point across is predicting cancer in patients. What do you think is more harmful, a false positive or a false negative? Clearly it is much better to be safe than sorry and predicting that a patient has cancer when they don't actually is much better than predicting a patient is negative when they actually do have cancer.

There are a few calculations we can do with these True/False Positives/Negatives. Let's get into it.


**Precision – What percent of your predictions were correct?**

Precision is the ability of a classifier not to label an instance positive that is actually negative. For each class it is defined as the ratio of true positives to the sum of true and false positives.

- TP – True Positives

- FP – False Positives

Precision – Accuracy of positive predictions.

- Precision = TP/(TP + FP)

**Recall – What percent of the positive cases did you catch?**

Recall is the ability of a classifier to find all positive instances. You can think of it as the model's ability to select relevant cases. For each class it is defined as the ratio of true positives to the sum of true positives and false negatives.

- FN – False Negatives

Recall: Fraction of positives that were correctly identified.

- Recall = TP/(TP+FN)

**F1 score – What percent of positive predictions were correct?**

The F1 score is a weighted harmonic mean of precision and recall such that the best score is 1.0 and the worst is 0.0. Generally speaking, F1 scores are lower than accuracy measures as they embed precision and recall into their computation. As a rule of thumb, the weighted average of F1 should be used to compare classifier models, not global accuracy. Since F1 score is a combination of precision and recall, it is usually your best bet in terms of evaluating a classification model, but it is also important to investigate precision and recall to see where your model is off.

- F1 Score = 2*(Recall * Precision) / (Recall + Precision)

**Implementing K-Means**




Now that our data is ready we can begin the Kmeans process.

We can use something called a silhouette score to find the optimal number of clusters. You can read more about silhouette analysis here. Essentially, we are trying to find the highest number for the silhouette average and then assign that number as our K, or number of clusters. From wikipedia:

"The silhouette value is a measure of how similar an object is to its own cluster (cohesion) compared to other clusters (separation). The silhouette ranges from −1 to +1, where a high value indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters. If most objects have a high value, then the clustering configuration is appropriate. If many points have a low or negative value, then the clustering configuration may have too many or too few clusters."

We'll bring in a utility function from **sklearn** called **silhouette_score** that will get us an average for the silhouette score for each data point in our clusters, given a value for K. We are going to be using the K that maximizes this silhouette score.

````python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns; sns.set_style('whitegrid');
````

````python
avgs = []

"""
We are going iterate from k=4 -> k=34 and try to find the silhouette score for each value of K.
We are going to choose the K that results in the highest score.
"""
start = 4 
stop = 29

for n_clusters in range(start, stop):
    """
    We are clustering our data with the Avg. column as our one feature.
    """
    
    X = df[['Avg']].values
    
    """
    sklearn.cluster.KMeans documentation
    
    Notice that the process for fitting the KMeans model is more or less equivalent to the process
    we went through with the LinearRegression model we pulled out from the linear_model module.
    
    https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html
    """
    
    model = KMeans(n_clusters=n_clusters) # instantiate the model with hyperparameters
    
    model.fit(X) # fit the model
    
    labels = model.predict(X) # predict labels (assign clusters) to our data
    
    silhouette_avg = silhouette_score(X, labels) # calculate the silhouette avg for our labels
    
    avgs.append(silhouette_avg) # append it to a list for plotting later

"""
Plotting the results.
np.arange is very similar to Python's range function
numpy.arange documentation
https://numpy.org/doc/stable/reference/generated/numpy.arange.html
"""
plt.plot(np.arange(start, stop, 1), avgs);
plt.xlabel('Number of clusters');
plt.ylabel('Silhouette score');
plt.ylim(0.2,0.8);

"""
matplotlib.pyplot.xticks allows us to set the xticks manually.
https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.xticks.html
"""
plt.xticks(np.arange(start, stop, 1));
````

**Mapping Our Tiers** 

If you take a look at one of the **labels** objects we create in the loop above, you'll notice that the labels KMeans assigns are seemingly random. KMeans doesn't have a concept of "tiers". We need to map the random labels we get back from KMeans to tiers we can use in Fantasy Basketball analysis.

````python
pd.set_option('display.max_rows', None)

k = 10 # this is the value we got back from doing the silhouette analysis above

model = KMeans(n_clusters=k)

model.fit(df[['Avg']].values)

labels = model.predict(X)

def assign_tiers(labels):
    """
    This function maps the random labels to actual fantasy basketball tiers.
    """
    
    unique_labels = []
    tiers = []
    
    for i in labels:
        
        if i not in unique_labels:
            unique_labels.append(i)
        
        """
        The Python set object gets us back a list-like object with only the unique values.
        The way this function works is that at each iteration, unique_labels is appended with the label.
        The length of the set is the tier we are on. Each time we get another unique label,
        the length of the set increases by one, and we now know we are on the next tier.
        
        
        
        """
        tiers.append(
            len(set(unique_labels))
        )
        
    return tiers

# map labels -> tiers
tiers = assign_tiers(labels)

# create a new column for tiers
df['Tier'] = tiers

df.set_index('Tier').head(100)
````

````python
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns

plt.style.use('fivethirtyeight')

def make_clustering_viz(df, figsize=(10, 10)):
    
    """
    Create a clustering visualization given our df.
    
    """
    
    # some of these values are hexadecimal color codes.
    # https://htmlcolorcodes.com/
    colors = ['purple', 'magenta', 'red', 'blue', 'orange', 'green',
              'salmon', 'yellow', 'black', 'grey', '#3498db', '#16a085', '#f4d03f', '#f1948a',
              '#48c9b0', '#3498db', '#e74c3c', '#d7bde2', '#d0d3d4'
              ]
    
    # create a dictionary with the keys of 1 -> 5, and the corresponding color for each tier.
    colors = dict(zip(range(1, k+1), colors[:k]))
    
    # set the figure size
    plt.figure(figsize=figsize)
    
    # plot Avg. vs. Rank. Our visualization will have Avg on the x-axis and ECR on the y-axis.
    plt.scatter(x=df['Avg'], 
                y=df['Rank'], 
                c='grey', # color of the point
                alpha=0.9, # 90% transparency
                s=50, # size 
                zorder=1000
               )
    
    yticks = [] # for storing player-names on the y-axis
    
    for _, row in df.iterrows():
        
        # we are going to plot a horizontal line along each point with a range stemming from
        # best (xmin) to worst (xmax) ranking.
        
        xmin = row['Best'] 
        xmax = row['Worst']
        
        ymin, ymax = row['Rank'], row['Rank'] 
        
        player = row['Player'] # annotate the y-ticks with player names
        
        tier = row['Tier']
        
        # plot a horizontal line
        plt.plot((xmin, xmax), (ymin, ymax), c=colors.get(tier, 'black'), alpha=0.8) # plot each individual horizontal line with either the tier color value or black
        
        yticks.append(player) # append the yticks list with the player's name
    
    # this will be used for the legend
    # for more info on legend patches, visit
    # https://matplotlib.org/3.3.0/api/_as_gen/matplotlib.patches.Patch.html#matplotlib.patches.Patch
    patches = []
    for tier, color in colors.items():
        patch = mpatches.Patch(color=color, label=f'Tier {tier}') # legend label + color
        patches.append(patch)
    
    plt.legend(handles=patches, borderpad=1, fontsize=12) # update the legend with the patches
    
    # setting xlabel, ylabel and yticks (player names)
    plt.xlabel('Average Expert Rank', fontsize=12)
    plt.ylabel('Expert Consensus Rank', fontsize=12)
    plt.yticks(df['Rank'], yticks, fontsize=12)
    
    # setting title
    plt.title('Tiers for the 2021 Draft. ECR vs. Average Expert Rank.', fontsize=12)
    
    # gca stands for get current axis. It's a roundabout way of grabbing the ax object like we
    # would if we were using matplotlib's object-oriented API.
    # docs on gca
    # https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.gca.html
    # we are inverting the y-axis here to show top players at the top.
    # docs on Axes.invert_yaxis()
    # https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.invert_yaxis.html
    
    plt.gca().invert_yaxis()
    plt.show()
````

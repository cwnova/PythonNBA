```python

import pandas as pd
from nba_api.stats.endpoints import PlayerGameLogs

# Set the season and minimum points threshold
season = '2019-20'
threshold = 35

# Fetch player game logs for the specified season
player_logs = PlayerGameLogs(season_nullable=season)

# Convert the player logs data to a DataFrame
logs_df = player_logs.get_data_frames()[0]

# Filter the DataFrame to include only games where the player scored at least the threshold number of points
high_scorers = logs_df[logs_df['PTS'] >= threshold]

# Display the high scorers DataFrame
# print(high_scorers[['GAME_DATE', 'PLAYER_NAME', 'PTS']])

# Sort the DataFrame by player name
high_scorers_sorted = high_scorers.sort_values(by='PLAYER_NAME')

# Display the sorted high scorers DataFrame
# print(high_scorers_sorted[['GAME_DATE', 'PLAYER_NAME', 'PTS']])

# Calculate the total number of 35+ point games for each player
total_games_19_20 = high_scorers_sorted.groupby('PLAYER_NAME').size().reset_index(name='Total Games')

# Display the total number of 35+ point games for each player
# print(total_games)

# Add a column for the regular season value
total_games_19_20['Regular Season'] = season

# Display the updated total_games DataFrame
# print(total_games_19_20)
#total_games.head()
```

```python
# combine data sets for 19-20 season through 2022-2023 season
combined_df = pd.concat([total_games_19_20, total_games_20_21,total_games_21_22, total_games_22_23 ])

# Display the combined DataFrame
print(combined_df)
```

```python
# Group the combined_df by Player Name and calculate the sum of Total Games for each player
summed_df = combined_df.groupby('PLAYER_NAME')['Total Games'].sum().reset_index()

# Sort the DataFrame by Total Games in descending order
summed_df_sorted = summed_df.sort_values('Total Games', ascending=True)

min_total_games = 10
filtered_df = summed_df_sorted[summed_df_sorted['Total Games'] >= min_total_games]

# Display the summed_df DataFrame
#print(summed_df)
#print(summed_df_sorted)
print(filtered_df)
```

```python
# Create a color library where Jimmy Butler is red and any other player is grey

import numpy as np

# Define the highlight color
highlight_color = '#98002E' ##'red'

# Create an empty list to store the colors
colors = []

# Iterate over the values in the 'PLAYER_NAME' column
for player_name in filtered_df['PLAYER_NAME']:
    # Check if the player name is 'Jimmy Butler'
    if player_name == 'Jimmy Butler':
        colors.append(highlight_color)
    else:
        colors.append('grey')

# Convert the colors list to a NumPy array
colors = np.array(colors)
```

```python
import matplotlib.pyplot as plt

# Extract the last name from the 'PLAYER_NAME' column
filtered_df['Last Name'] = filtered_df['PLAYER_NAME'].str.split().str[-1]

# Create a bar plot with filtered data and set color to colors library, which has Jimmy Butler = red
fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(filtered_df['Last Name'], filtered_df['Total Games'], color=colors)  # Assigning colors library from above
plt.xticks(rotation=90)
plt.xlabel('Player')  # Update the x-axis label to 'Last Name'
plt.ylabel('Total Games')
plt.title('Who has scored the most 35+ point games in the NBA Regular Season?')
subtitle = '2019-2020 through 2022-2023 Season'
plt.text(0.5, .95, subtitle, transform=ax.transAxes, ha='center')

# Add value labels on top of selected bars
for i, (last_name, total_games) in enumerate(zip(filtered_df['Last Name'], filtered_df['Total Games'])):
    if last_name in ['Butler']:
        ax.text(i, total_games, str(total_games), ha='center', va='bottom')

plt.show()
```
